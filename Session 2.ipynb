### 1.6 What is really broadcasting ?

Broadcasting is the ability to perform computations on or using arrays without the need to use for loops even if arrays were not of the same exact shape

```python
lst1=[1,2,3]
lst2=[1,1,1]
```

```python
## if we try to add lst1 and lst2 it just concats the lists together
lst1+lst2
```

```python
# if we try to multiply lst1 by 2 , it just repeats the list twice
2*lst1
```

```python
## now what if we want to actually peform a computation?

## adding lst1 items to lst2 
lst_new=[]
for i in range(len(lst2)):
    lst_new.append(lst2[i]+lst1[i])
```

```python
lst_new
```

```python
## adding multiply lst1 by 2
lst_new=[]
for i in range(len(lst1)):
    lst_new.append(lst1[i]*2)
```

```python
lst_new
```

in numpy arrays,things are different , broadcast is done impliciltly and thus compuations are much faster and easier as well

### Broadcasting Rules:
*Broadcasting two arrays together follow these rules:*
 

* If the arrays don’t have the same rank then prepend the shape of the lower rank array with 1s until both shapes have the same length.
* The two arrays are compatible in a dimension if they have the same size in the dimension or if one of the arrays has size 1 in that dimension.
* The arrays can be broadcast together iff they are compatible with all dimensions.
* In any dimension where one array had size 1 and the other array had size greater than 1, the first array behaves as if it were copied along that dimension.

(2,2,1), (1,1,2) ---> not compatible 

```python
a=np.ones((2,3))
b=np.zeros((5,6))
```
